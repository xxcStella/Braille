```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = [0, 0.2, 0.3, 0.5]
depth = [0, 2, 2, 0]
plt.plot(t, depth, label='depth: 2mm')
plt.xlabel('time [s]')
plt.ylabel('depth [mm]')
plt.title('TacTip vertical moving trajactory')
plt.text(0.05, 0.25, 'pressing speed: 10mm/s')
plt.text(0.19, 1.75, 'hold phase: 0.1s')
plt.legend()
```
